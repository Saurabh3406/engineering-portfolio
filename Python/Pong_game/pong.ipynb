```python
import pygame as pg

##variables
#screen dimensions
LENGTH = 800
HEIGHT = 600
#boundary width
WIDTH = 10



#ball speed
BALL_SPEED = 4
#ball radius
BALL_RADIUS = 10

#colors
WHITE = pg.Color('white')
BLACK = pg.Color('black')
RED = pg.Color('red')
GREEN = pg.Color('green')
BLUE = pg.Color('blue')

#paddle dimensions
PADDLE_WIDTH = 10
PADDLE_HEIGHT = 50

#Scores of the players
player1_score = 0
player2_score = 0

#classes
class Paddle:
    def __init__(self, y):
        self.y = y
    
    def show(self, COLOUR):
        global screen, PADDLE_WIDTH, PADDLE_HEIGHT
        pg.draw.rect(screen, COLOUR, (0, self.y, PADDLE_WIDTH, PADDLE_HEIGHT))
    
    def move(self):
        self.show(BLACK)
        self.y = pg.key.get_pressed()[pg.K_UP] * -5 + pg.key.get_pressed()[pg.K_DOWN] * 5 + self.y
        # Ensure the paddle stays within the screen bounds
        if self.y < WIDTH:
            self.y = WIDTH 
        elif self.y > HEIGHT - PADDLE_HEIGHT - WIDTH:
            self.y = HEIGHT - PADDLE_HEIGHT - WIDTH
        self.show(GREEN)

class Paddle2:
    def __init__(self, y):
        self.y = y
    
    def show(self, COLOUR):
        global screen, PADDLE_WIDTH, PADDLE_HEIGHT
        pg.draw.rect(screen, COLOUR, (LENGTH - PADDLE_WIDTH, self.y, PADDLE_WIDTH, PADDLE_HEIGHT))
    
    def move(self):
        self.show(BLACK)
        self.y = pg.mouse.get_pos()[1]
        # Ensure the paddle stays within the screen bounds
        if self.y < WIDTH:
            self.y = WIDTH 
        elif self.y > HEIGHT - PADDLE_HEIGHT - WIDTH:
            self.y = HEIGHT - PADDLE_HEIGHT - WIDTH
        self.show(GREEN)

class Ball:

    def __init__(self, x, y, vx, vy):
        BALL_POSITION_X = LENGTH//2
        BALL_POSITION_Y = HEIGHT//2
        #ball speed
        global BALL_SPEED, BALL_SPEED
        self.vx = -BALL_SPEED
        self.vy = BALL_SPEED
        self.x = BALL_POSITION_X
        self.y = BALL_POSITION_Y

    def show(self, COLOUR):
        global screen, BALL_RADIUS
        pg.draw.circle(screen, COLOUR, (self.x, self.y), BALL_RADIUS)

    def move(self):
        global BLACK, RED
        self.show(BLACK)
        self.x += self.vx
        self.y += self.vy
        self.show(RED)

# This script initializes a Pygame window and draws a white boundary on a black background.
# It runs at a frame rate of 60 FPS and listens for the quit event to close the window.

#frame rate
FPS = 60

#initialize pygame
pg.init()

# Create a font object
font = pg.font.SysFont(None, 36)

#initialize the ball and paddles
#ball starts at the center of the screen with a speed of BALL_SPEED in both x and y directions
ballplay = Ball(LENGTH//2, HEIGHT//2, BALL_SPEED, BALL_SPEED)
paddleplay = Paddle(HEIGHT // 2 - PADDLE_HEIGHT // 2)
paddleplay2 = Paddle2(HEIGHT // 2 - PADDLE_HEIGHT // 2)

#initialize the screen
screen = pg.display.set_mode((LENGTH, HEIGHT))
pg.display.set_caption("Pong Game")

#fill the screen with black
screen.fill(BLACK)

#draw the two boundary lines
pg.draw.rect(screen, WHITE, (0, 0, LENGTH, WIDTH))
pg.draw.rect(screen, WHITE, (0, HEIGHT-WIDTH, LENGTH, WIDTH))

#draw the players' paddles
paddleplay.show(GREEN)
paddleplay2.show(GREEN)

#draw the ball at the center of the screen
ballplay.show(RED)

#initialize the clock
clock = pg.time.Clock()

#main loop
running = True
while running:
    #event handling
    for event in pg.event.get():
        if event.type == pg.QUIT:
            running = False

    #check for collisions with the boundaries
    if ballplay.y <= WIDTH + BALL_RADIUS or ballplay.y >= HEIGHT - WIDTH - BALL_RADIUS:
        ballplay.vy = -ballplay.vy
    
    #check for collisions with both the paddles
    if (ballplay.x == PADDLE_WIDTH + BALL_RADIUS and 
        paddleplay.y <= ballplay.y <= paddleplay.y + PADDLE_HEIGHT):
        ballplay.vx = -ballplay.vx
    elif (ballplay.x == LENGTH - PADDLE_WIDTH - BALL_RADIUS and 
          paddleplay2.y <= ballplay.y <= paddleplay2.y + PADDLE_HEIGHT):
        ballplay.vx = -ballplay.vx
    
    #show initial scores on the screen
    score_text = font.render(f"Player 1: {player1_score}  Player 2: {player2_score}", True, WHITE)
    screen.blit(score_text, (LENGTH // 2 - score_text.get_width() // 2, 10))
    #check if the ball goes out of bounds
    if ballplay.x < 0 or ballplay.x > LENGTH:
        #clear the lost ball from the screen
        ballplay.show(BLACK)
        
        #update the scores
        if ballplay.x < 0:
            player2_score += 1
        else:
            player1_score += 1

        #reset the ball position to the center
        ballplay.x = LENGTH // 2
        ballplay.y = HEIGHT // 2

        #reverse the ball's direction
        ballplay.vx = -ballplay.vx

        #fill the screen with black
        screen.fill(BLACK)

        #draw the boundary lines again
        #draw the two boundary lines
        pg.draw.rect(screen, WHITE, (0, 0, LENGTH, WIDTH))
        pg.draw.rect(screen, WHITE, (0, HEIGHT-WIDTH, LENGTH, WIDTH))
        
    pg.display.flip()
    ballplay.move()
    paddleplay.move()
    paddleplay2.move()
    
    #update the scores on the screen
    score_text = font.render(f"Player 1: {player1_score}  Player 2: {player2_score}", True, WHITE)
    screen.blit(score_text, (LENGTH // 2 - score_text.get_width() // 2, 10))

    #cap the frame rate
    clock.tick(FPS)
```
